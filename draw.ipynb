```python
import matplotlib.pyplot as plt

# ====== 准备数据（示例）======
# 上面一行：Spatial ablation（横坐标 4 个）
x_spatial = ['(6,6,6)', '(12,12,12)', '(18,18,18)', '(24,24,24)']
spatial_acc_1 = [64.95, 65.98, 63.92, 61.86]   # 左上子图数据（替换为你的数据）
spatial_acc_2 = [62.89, 64.95, 61.86, 60.82]   # 右上子图数据（替换为你的数据）

# 下面一行：Temporal ablation（横坐标 3 个）
x_temporal = ['6s', '9s', '12s']
temporal_acc_1 = [64.95, 62.89, 61.86]            # 左下子图数据（替换为你的数据）
temporal_acc_2 = [65.98, 64.95, 61.86]            # 右下子图数据（替换为你的数据）

# ====== 2×2 子图 ======
fig, axes = plt.subplots(2, 2, figsize=(10, 8), sharey=True)

# ---- 上排：Spatial ablation ----
ax = axes[0, 0]
ax.plot(x_spatial, spatial_acc_1, '-o', linewidth=2, markersize=7)
ax.set_title('Spatial ablation (temporal duration=6s)', fontsize=12, weight='bold')
ax.set_xlabel('Resolution (mm per side)', fontsize=10)
ax.set_ylabel('Accuracy (%)', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylim(60, 68)

ax = axes[0, 1]
ax.plot(x_spatial, spatial_acc_2, '-o', linewidth=2, markersize=7)
ax.set_title('Spatial ablation (temporal duration=12s)', fontsize=12, weight='bold')
ax.set_xlabel('Resolution (mm per side)', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylim(60, 68)

# ---- 下排：Temporal ablation ----
ax = axes[1, 0]
ax.plot(x_temporal, temporal_acc_1, '-o', linewidth=2, markersize=7)
ax.set_title('Temporal ablation (spatial voxel=6mm)', fontsize=12, weight='bold')
ax.set_xlabel('Temporal duration (s)', fontsize=10)
ax.set_ylabel('Accuracy (%)', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylim(60, 68)

ax = axes[1, 1]
ax.plot(x_temporal, temporal_acc_2, '-o', linewidth=2, markersize=7)
ax.set_title('Temporal ablation (spatial voxel=12mm)', fontsize=12, weight='bold')
ax.set_xlabel('Temporal duration (s)', fontsize=10)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylim(60, 68)

# 全局标题与布局
fig.suptitle('Ablation Study: Spatial (top) vs Temporal (bottom)', fontsize=16, weight='bold')
fig.tight_layout(rect=[0, 0, 1, 0.95])

fig.savefig("ablation_spatial_temporal.svg", format="svg", bbox_inches="tight")
plt.show()

plt.show()


```
